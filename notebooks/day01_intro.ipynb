# 1. What is a Computer?

## Q1. What is a computer?

**Answer:**

A computer is an electronic machine that receives data, processes it according to given instructions, and produces meaningful output.

A computer is extremely fast but does not think like humans. It only follows precise instructions.

---

## Q2. Why do computers need exact instructions?

**Answer:**

Computers cannot understand vague instructions or make assumptions like humans.

For example, a human understands:

"Add some salt to the food."

A computer would need exact details such as:

* How much salt should be added?
* Which food should receive the salt?
* At what stage should the salt be added?

Therefore, computer instructions must be:

* Clear
* Ordered
* Unambiguous

---

# 2. What is Programming?

## Q3. What is a program?

**Answer:**

A program is a collection of instructions written in a programming language that tells a computer what tasks to perform.

Example:

A calculator application is a program that receives numbers, performs calculations, and displays the result.

---

## Q4. What is programming?

**Answer:**

Programming is the process of designing and writing instructions that tell a computer how to solve a problem.

---

# 3. What is an Algorithm?

## Q5. What is an algorithm?

**Answer:**

An algorithm is a step-by-step procedure for solving a problem or completing a task.

It describes the logic of a solution and is independent of any programming language.

Example: Finding the average marks of three students.

Algorithm:

1. Take three marks.
2. Add all marks.
3. Divide the total by three.
4. Display the average.

---

## Q6. What is the difference between an algorithm and a program?

**Answer:**

| Algorithm                               | Program                                                 |
| --------------------------------------- | ------------------------------------------------------- |
| A logical plan to solve a problem       | A set of instructions written in a programming language |
| Independent of any programming language | Written using a specific language                       |
| Describes the steps required            | Tells the computer how to execute those steps           |

# 5. What is Data Science?

## Q9. What is Data Science?

**Answer:**

Data Science is the field that combines programming, mathematics, statistics, and domain knowledge to extract useful information and insights from data.

---

## Q10. What types of questions does Data Science answer?

**Answer:**

### 1. What happened? (Descriptive Analysis)

Example:

Sales decreased by 20% last month.

---

### 2. Why did it happen? (Diagnostic Analysis)

Example:

Customers stopped buying because the price became too high.

---

### 3. What will happen? (Predictive Analysis)

Example:

Sales may continue decreasing next month.

---

### 4. What should be done? (Prescriptive Analysis)

Example:

Offer discounts or change the pricing strategy.

---

# 6. How Does a Computer Work?

## Q11. What is the basic working cycle of a computer?

**Answer:**

A computer follows the IPO cycle:

Input → Processing → Output → Storage

---

## Q12. What is Input?

**Answer:**

Input is the data or instructions provided to a computer.

Examples:

* Keyboard
* Mouse
* Camera
* Sensors
* Internet data

---

## Q13. What is Processing?

**Answer:**

Processing is the stage where a computer performs calculations and makes decisions based on instructions.

The CPU (Central Processing Unit) is mainly responsible for processing.

---

## Q14. What is Output?

**Answer:**

Output is the final result produced by a computer.

Examples:

* Numbers displayed on the screen
* Graphs
* Recommendations from applications

---

## Q15. What is Storage?

**Answer:**

Storage is where information is saved for future use.

Examples:

* SSD
* Hard disk
* Database
* Cloud storage

---

# 7. How Do Computers Understand Information?

## Q16. What language does a computer understand?

**Answer:**

At the lowest level, computers understand binary language.

Binary consists of only two digits:

0 and 1

Where:

* 0 represents OFF
* 1 represents ON

These combinations of zeros and ones represent numbers, text, images, videos, and programs.

---

# 8. Why Do We Use Programming Languages?

## Q17. Why don't humans write programs directly in binary?

**Answer:**

Binary is difficult for humans to read and write.

Therefore, we use high-level programming languages such as:

* Python
* C++
* Java
* JavaScript

These languages are translated into machine instructions that computers can understand.

---

# 9. Why Python for Data Science?

## Q18. Why is Python widely used in Data Science?

**Answer:**

Python is popular because it is:

* Easy to learn
* Easy to read
* Powerful and flexible
* Supported by a large community

Important Python libraries used in Data Science:

* NumPy → Numerical computation
* Pandas → Data manipulation and analysis
* Matplotlib → Data visualization
* Scikit-learn → Machine Learning
* TensorFlow and PyTorch → Deep Learning

---

# 10. Key Points to Remember

* Computers follow instructions; they do not think.
* A program is a set of instructions written in a programming language.
* An algorithm is a logical step-by-step solution.
* Data is raw information collected from the world.
* Data Science converts data into useful knowledge.
* Computers follow the IPO cycle.
* Computers understand binary language.
